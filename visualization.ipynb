## 3. Rotate pc

### 3.1 Generate Rotated data

```python
### pre-code
import os
import sys
sys.path.append(os.path.join('.','data_preprocessing'))
import arch_dataloader
import pre_rotation
import numpy as np
import h5py
import argparse
import numpy as np
from datetime import datetime

# read data
split='train'
arch_data_dir = './data/arch3_no_others_combined_1.5m_2048'
train_filelist = os.path.join(arch_data_dir, "train_data_files.txt")
train_dataset = arch_dataloader.get_dataloader(filelist=train_filelist, num_points=2048, num_dims=3, split='train', batch_size=32, num_workers=4, drop_last=True)
print("classifer set size: " + str(train_dataset.dataset.__len__()))

file_size = 2048
batch_size = 128
point_num = 2048
NUM_CLASSES = 4
batch_num = file_size / batch_size
rotated_dataset = np.zeros((file_size, point_num, 3))
rotated_labelset = np.zeros((file_size), dtype=np.int32)
idx = 0
idx_h5 = 0
sample_name = 'combined_1.5m'
# rotate data 
for iter, data in enumerate(train_dataset):
    input_points, _ = data
    rotated_data, rotated_label = pre_rotation.rotate_pc(current_data=input_points, NUM_CLASSES=4)
    start = idx*batch_size
    end = (idx+1)*batch_size
    rotated_dataset[start:end, ...] = rotated_data
    rotated_labelset[start:end, ...] = rotated_label
    idx += 1
    if (idx == batch_num):
        filename_h5 = os.path.join('./data', 'rotated_%d_angle_%s_%d' % (NUM_CLASSES, sample_name, point_num), '%d.h5' % idx_h5)
        print('{}-Saving {}...'.format(datetime.now(), filename_h5))
        file = h5py.File(filename_h5, 'w')
        file.create_dataset('data', data=rotated_dataset[0:end, ...])
        file.create_dataset('label', data=rotated_labelset[0:end, ...])
        file.close()
        idx_h5 +=1
        idx = 0
```

### 3.2 Test & Visualize Rotated PC

```python
import matplotlib.pyplot as plt
def pyplot_draw_point_cloud(points):
    """ points is a Nx3 numpy array """
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(points[:,0], points[:,1], points[:,2])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    fig.show()
```
